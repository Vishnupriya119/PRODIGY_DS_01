```python
plt.figure(figsize=(10, 6))
plt.bar(df['Country Name'], df['2021'])
plt.title('Population Distribution by Country in 2021')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45) 
plt.show()
```

```python
for i in years:
    fig=plt.figure(figsize=(3,3))
    plt.hist(df[i],color='yellow',bins=10)
    plt.xlabel(i)
    plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.hist(df['2021'], bins=10, color='skyblue')
plt.title('Population Distribution in 2021')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
barWidth = 2.5
r1 = range(len(df['Country Name']))
r2 = [x + barWidth for x in r1]
plt.bar(r1, df['2020'], color='b', width=barWidth, edgecolor='grey', label='2020')
plt.bar(r2, df['2021'], color='y', width=barWidth, edgecolor='grey', label='2021')
plt.xlabel('Country', fontweight='bold')
plt.title('Population Comparison by Year')
plt.legend()
plt.show()
```

```python
selected_year = '2020' 
data = df[['Country Name', selected_year]]
plt.figure(figsize=(15, 8))
colors = sns.color_palette("husl", n_colors=len(data))
bars = plt.bar(data['Country Name'], data[selected_year], color=colors)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Distribution of Population in {selected_year}')
```

```python
selected_countries = ['China', 'India', 'United States', 'Brazil', 'Russia']
data = df[df['Country Name'].isin(selected_countries)].set_index('Country Name').T
data = data.T
plt.figure(figsize=(15, 8))
data.plot(kind='bar', cmap='viridis', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Comparison of Selected Countries Over Time')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
ddf = df[['Country Name', '2020']]
df = df.sort_values(by='2020', ascending=False)
plt.figure(figsize=(10, 8))
plt.barh(df['Country Name'].head(10), df['2020'].head(10), color='m')
plt.xlabel('Population Size')
plt.ylabel('Country')
plt.title('Top 10 Countries by Population Size in 2020')
plt.show()
```

```python
df = df[['Country Name', '2020']]
df = df.sort_values(by='2020', ascending=False)
colors = plt.cm.viridis(np.linspace(0, 1, 10))  
plt.figure(figsize=(12, 6))
bars = plt.bar(df['Country Name'][:10], df['2020'][:10], color=colors)
plt.xlabel('Country')
plt.ylabel('Population Size')
plt.title('Top 10 Countries by Population Size in 2020')
plt.legend(bars, df['Country Name'][:10], loc='upper right', bbox_to_anchor=(1.25, 1))
plt.xticks(rotation=45)
plt.show()
```

```python
continent_mapping = {
    'China': 'Asia',
    'India': 'Asia',
    'United States': 'North America',
    'Brazil': 'Latin America & Caribbean',
    'Russia': 'Europe'
}
df['Continent'] = df['Country Name'].map(continent_mapping)
continent_population = df.groupby('Continent')['2020'].sum()
plt.figure(figsize=(10, 6))
colors = sns.color_palette("Set3", n_colors=len(continent_population))
continent_population.sort_values().plot(kind='bar', color=colors)
plt.xlabel('Continent')
plt.ylabel('Total Population')
plt.title('Total Population Distribution Across Continents in 2020')
plt.show()
```
